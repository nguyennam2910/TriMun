```python
def split_data(data_dict, keyname, ind_list):
    name_number = [37, 43, 45, 49, 51, 67, 68, 69, 72, 75, 76, 82, 84, 85, 87, 89, 91, 93, 94, 99]
    data_list = []
    t = 0
    if keyname=='RESPr_':
        for i, number in enumerate(name_number[1:]):
            name = keyname + str(number)
            data = data_dict[name]
            for j, ii in enumerate(ind_list[i]):
                if j == 0:
                    jj = ii
                else:
                    t+=1
                    day_data = data[jj:ii]
                    jj = ii
                    data_list.append(day_data)
    elif keyname == 'GSR_':
        minG = []
        maxG = []
        for i, number in enumerate(name_number):
            
            name = keyname + str(number)
            data = data_dict[name]
            minG.append(np.min(data))
            maxG.append(np.max(data))
            if i >7 and i < 10:
                print(data.shape)
            # ct tn hiu theo index 
            for j, ii in enumerate(ind_list[i]):
                if j == 0:
                    jj = ii
                else:
                    t+=1
                    day_data = data[jj:ii]
                    jj = ii
                    data_list.append(day_data)
        return data_list, minG, maxG
    else:
        for i, number in enumerate(name_number):
            name = keyname + str(number)
            data = data_dict[name]
            if i >7 and i < 10:
                print(data.shape)
            # ct tn hiu theo index 
            for j, ii in enumerate(ind_list[i]):
                if j == 0:
                    jj = ii
                else:
                    t+=1
                    day_data = data[jj:ii]
                    jj = ii
                    data_list.append(day_data)
    print(t)
    return data_list


        
```
