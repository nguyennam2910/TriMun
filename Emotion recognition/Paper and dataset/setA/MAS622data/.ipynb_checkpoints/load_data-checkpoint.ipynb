### Sắp xếp lại dữ liệu theo thứ tự cảm xúc: 

```python
def data_Create(day):
    dict = {}
    data = {}

    for i in range(0,32):
        if i%8 == 0:
            t1 = "No Emotion"
        elif i%8 == 1:
            t1 = "Anger"
        elif i%8 == 2:
            t1 = "Hate"
        elif i%8 == 3:
            t1 = "Grief"
        elif i%8 == 4:
            t1 = "P-Love"
        elif i%8 == 5:
            t1 = "R-Love"
        elif i%8 == 6:
            t1 = "Joy"
        else:
            t1 = "Reverence"
        if i/8 < 1:
            t2 = "EMG(jaw)"
            ymin = -10
            ymax = 200
        elif i/8 >=1 and i/8 <2:
            t2 = "BVP"
            ymin = 0
            ymax = 100
        elif i/8 >=2 and i/8 <3:
            t2 = "GSR(palm)"
            ymin = 0
            ymax = 5
        else:
            t2 = "Respiration"
            ymin = 35
            ymax = 70

#         t = [t1, t2]  
        dict[t1] = day[:, i]
        if i%8 == 7:
            data[t2] = dict
            dict = {}
    return data
```

```python
i = 1
alldays_data = {}
for day in days:
    data = data_Create(day)
    alldays_data["day" +str(i)] = data
    i += 1
print(alldays_data.keys())
```
